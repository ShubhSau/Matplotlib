Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Scatter Plot: Scatter plots are used to visualize the relationship between two continuous variables. It is a diagram where each value in the data set is represented by a dot.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

A3: The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to plot multiple plots in separate sub-areas of a larger figure, enabling the visualization of multiple datasets or different aspects of data side by side.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line Plot 4')

plt.tight_layout()

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Bar Plot: A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

It is Used for:
1. Comparison of Categories
2. Data Distribution
3. Categorical Relationships
4. Presentation and Communication

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Box Plot: It is a type of statistical visualization that displays the distribution of a dataset through five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It is used to identify the central tendency, spread, and skewness of the data, as well as to detect outliers.

```python
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)
```

```python
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)
```
